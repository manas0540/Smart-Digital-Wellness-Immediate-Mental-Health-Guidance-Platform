SKWENESS CHECK

```python
import pandas as pd
import numpy as np

# numeric columns
num_cols = df.select_dtypes(include=[np.number]).columns.tolist()

# skewness series
skewness = df[num_cols].skew().sort_values(ascending=False)
skewness_df = pd.DataFrame({'skewness': skewness})
skewness_df
```

SKEWNESS DEFINING

```python
def interpret_skew(s):
    a = abs(s)
    if a < 0.5:
        return "approx symmetric"
    if a < 1:
        return "moderately skewed"
    return "highly skewed"

skewness_df['interpretation'] = skewness_df['skewness'].apply(interpret_skew)
skewness_df
```

```python


def plot_distribution(col):
    plt.figure(figsize=(8,4))
    sns.histplot(df[col].dropna(), kde=True)
    plt.title(f"{col} â€” skewness = {df[col].skew():.3f}")
    plt.xlabel(col)
    plt.ylabel("count")
    plt.show()

# Example: plot for top 3 most skewed numeric cols
for c in skewness_df.index[:3]:
    plot_distribution(c)
```

```python
from scipy import stats

col = skewness_df.index[0]
series = df[col].dropna()
if (series <= 0).any():
    print("Box-Cox requires positive values â€” skip or shift the data first.")
else:
    transformed, lam = stats.boxcox(series)
    print("Box-Cox lambda:", lam)
    # put back into dataframe aligned by index if needed
    df.loc[series.index, f"{col}_boxcox"] = transformed
    print("skew after boxcox:", pd.Series(transformed).skew())
    plot_distribution(f"{col}_boxcox")
```

```python
from sklearn.preprocessing import PowerTransformer

pt = PowerTransformer(method='yeo-johnson', standardize=False)

# Pick skewed column
col = skewness_df.index[0]

# Handle NaNs
series = df[[col]].fillna(df[col].median())

# Apply Yeo-Johnson transform
transformed = pt.fit_transform(series)

# Add new column
df[f"{col}_yeojohnson"] = transformed

# Print skewness before/after
print("skew before:", df[col].skew())
print("skew after yeojohnson:", df[f"{col}_yeojohnson"].skew())

# Corrected call â†’ pass column name
plot_distribution(f"{col}_yeojohnson")
```

Univariate Analysis

```python




numeric_cols = ['Age','Daily_Screen_Time(hrs)','Sleeping_hours_avg','Stress_Level(1-10)','Happiness_Index(1-10)']

for col in numeric_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], bins=10, kde=True, color='skyblue')
    plt.title(f'{col} Distribution')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()


categorical_cols = ['Gender','Mood_Label','Excessive_used_app']

for col in categorical_cols:
    plt.figure(figsize=(6,4))
    sns.countplot(x=col, data=df, color='skyblue')  # use color instead of palette for single hue
    plt.title(f'{col} Count')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()

```

```python
#Numeric vs Numeric (Scatter plots)

plt.figure(figsize=(6,4))
sns.scatterplot(x='Daily_Screen_Time(hrs)', y='Happiness_Index(1-10)',
                hue='Mood_Label', data=df, palette='bright', s=100)
plt.title('Screen Time vs Happiness')
plt.xlabel('Daily Screen Time (hrs)')
plt.ylabel('Happiness Index')
plt.show()

plt.figure(figsize=(6,4))
sns.scatterplot(x='Stress_Level(1-10)', y='Sleeping_hours_avg',
                hue='Gender', data=df, palette='coolwarm', s=100)
plt.title('Stress Level vs Sleeping Hours')
plt.xlabel('Stress Level')
plt.ylabel('Sleeping Hours')
plt.show()

# Categorical vs Numeric (Boxplots)

plt.figure(figsize=(6,4))
sns.boxplot(x='Gender', y='Daily_Screen_Time(hrs)', data=df, color='lightgreen')
plt.title('Daily Screen Time by Gender')
plt.xlabel('Gender')
plt.ylabel('Daily Screen Time (hrs)')
plt.show()

plt.figure(figsize=(6,4))
sns.boxplot(x='Mood_Label', y='Happiness_Index(1-10)', data=df, color='lightcoral')
plt.title('Happiness Index by Mood Label')
plt.xlabel('Mood Label')
plt.ylabel('Happiness Index')
plt.show()

# Categorical vs Categorical (Countplot with hue)

plt.figure(figsize=(6,4))
sns.countplot(x='Excessive_used_app', hue='Mood_Label', data=df, palette='Set1')
plt.title('App Usage vs Mood Label')
plt.xlabel('Excessive Used App')
plt.ylabel('Count')
plt.show()



```
